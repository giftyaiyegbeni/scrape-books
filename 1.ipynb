### 1. Scrape books from http://books.toscrape.com (Book name, price, stock status (in stock or out of stock), rating, description, product information, category (poetry, fiction, historical fiction, etc)

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.keys import Keys
from random import shuffle
```

```python
# Configure Selenium webdriver
path = "C:\Program Files (x86)\Selenium\chromedriver.exe"

driver = webdriver.Chrome(path)

# Open the website
driver.get("http://books.toscrape.com")

# Wait for the books to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "product_pod")))
```

```python
# Get all the book elements
books = driver.find_elements(By.CLASS_NAME, "product_pod")
```

```python
for book in books:
    # Extract book information
    title = book.find_element(By.CSS_SELECTOR, "h3 > a").get_attribute("title")
    price = book.find_element(By.CLASS_NAME, "price_color").text
    stock_status = "In stock" if "availability" in book.find_element(By.CLASS_NAME, "availability").text.lower() else "Out of stock"
    rating = book.find_element(By.CLASS_NAME, "star-rating").get_attribute("class").split()[-1]
    
    # Open the book details page
    book_link = book.find_element(By.CSS_SELECTOR, "h3 > a")
    book_link.click()

    # Wait for the book details page to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "product_page")))

    # Extract additional information from the book details page
    description = driver.find_element(By.CSS_SELECTOR, "#product_description ~ p").text
    product_information = driver.find_element(By.CSS_SELECTOR, ".table-striped").text
    category = driver.find_element(By.CSS_SELECTOR, ".breadcrumb > li:nth-child(3) > a").text

    # Print the book information
    print("Title:", title)
    print("Price:", price)
    print("Stock Status:", stock_status)
    print("Rating:", rating)
    print("Description:", description)
    print("Product Information:", product_information)
    print("Category:", category)
    print("")

    # Go back to the main page
    driver.back()

# Close the browser
driver.quit()
```

### 2. Scrape the first 5 pages (20 books per page)

```python
# Configure Selenium webdriver
path = "C:\Program Files (x86)\Selenium\chromedriver.exe"

driver = webdriver.Chrome(path)

# Open the website
driver.get("http://books.toscrape.com")

# Wait for the books to load
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "product_pod")))
```

```python
# Scrape the first 5 pages
for page in range(1, 6):
    # Navigate to the page
    url = f"http://books.toscrape.com/catalogue/page-{page}.html"
    driver.get(url)

    # Wait for the books to load
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "product_pod")))

    # Get all the book elements
    books = driver.find_elements(By.CLASS_NAME, "product_pod")

    # Iterate over each book element
    for index, book in enumerate(books):
        # Extract book information
        title = book.find_element(By.CSS_SELECTOR, "h3 > a").get_attribute("title")
        price = book.find_element(By.CLASS_NAME, "price_color").text
        stock_status = "In stock" if "availability" in book.find_element(By.CLASS_NAME, "availability").text.lower() else "Out of stock"
        rating = book.find_element(By.CLASS_NAME, "star-rating").get_attribute("class").split()[-1]

        # Open the book details page
        book_link = book.find_element(By.CSS_SELECTOR, "h3 > a")
        book_link.click()

        # Wait for the book details page to load
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "product_page")))

        # Extract additional information from the book details page
        description = driver.find_element(By.CSS_SELECTOR, "#product_description ~ p").text
        product_information = driver.find_element(By.CSS_SELECTOR, ".table-striped").text
        category = driver.find_element(By.CSS_SELECTOR, ".breadcrumb > li:nth-child(3) > a").text

        # Print the book information
        print("Title:", title)
        print("Price:", price)
        print("Stock Status:", stock_status)
        print("Rating:", rating)
        print("Description:", description)
        print("Product Information:", product_information)
        print("Category:", category)
        print("")

        # Go back to the main page
        driver.back()

        # Re-locate the book elements on the main page after navigating back
        books = driver.find_elements(By.CLASS_NAME, "product_pod")

# Close the browser
driver.quit()
```

### 3. Scrape 10-20 distinct quote authors from http://quotes.toscrape.com (Name, nationality, description, date of birth)

```python
driver = webdriver.Chrome(path)
```

```python
quote_authors = set()

for i in range(1, 3):
    driver.get(f'https://quotes.toscrape.com/page/{i}/')
    quotes = driver.find_elements(By.CLASS_NAME, 'quote')
    print(f'Page number {i}: {len(quotes)} quotes')

    for quote in quotes:
        author_name = quote.find_element(By.CLASS_NAME, 'author').text
        if author_name not in quote_authors:
            author_bio_link = quote.find_element(By.CSS_SELECTOR, 'span a').get_attribute('href')

            driver.get(author_bio_link)

            author_nationality = driver.find_element(By.CLASS_NAME, 'author-born-location').text
            author_description = driver.find_element(By.CLASS_NAME, 'author-description').text.strip("“”")
            author_birth_date = driver.find_element(By.CLASS_NAME, 'author-born-date').text.strip("()")

            print("Author Name:", author_name)
            print("Nationality:", author_nationality)
            print("Description:", author_description)
            print("Date of Birth:", author_birth_date)
            print("")

            quote_authors.add(author_name)  # Append author to the set

            driver.back()

print("Distinct Quote Authors:")
for author in quote_authors:
    print(author)
```

### 4. Build a scraper that will scrape a random page from Wikipedia

```python
driver = webdriver.Chrome(path)

# Navigate to the Wikipedia homepage
driver.get("https://en.wikipedia.org")
```

```python
# Find the search input and submit a random term to trigger a random page
search_input = driver.find_element(By.ID, "searchInput")
search_input.send_keys("Special:Random")
search_input.submit()
```

```python

# Get the title and URL of the random article
article_title = driver.find_element(By.ID, "firstHeading").text
article_url = driver.current_url
```

```python
# Print the results
print("Random Article Title:", article_title)
print("Random Article URL:", article_url)

# Close the browser
driver.quit()
```
